```python
# Define the knowledge base (KB) for facts about fruits
knowledge_base = {
    'Apple': {
        'Color': ['Red', 'Green'],
        'Taste': 'Sweet and slightly tart',
        'Nutritional Value': 'Rich in Vitamin C and dietary fiber'
    },
    'Banana': {
        'Color': 'Yellow',
        'Taste': 'Sweet and creamy',
        'Nutritional Value': 'High in potassium and Vitamin B6'
    },
    'Orange': {
        'Color': 'Orange',
        'Taste': 'Sweet and citrusy',
        'Nutritional Value': 'Excellent source of Vitamin C'
    }
}

# Function to answer questions about fruits
def answer_question(fruit, question):
    if fruit in knowledge_base:
        if question == 'Color':
            return f"{fruit}'s color is {', '.join(knowledge_base[fruit]['Color'])}."
        elif question == 'Taste':
            return f"{fruit} tastes {knowledge_base[fruit]['Taste']}."
        elif question == 'Nutritional Value':
            return f"{fruit} is {knowledge_base[fruit]['Nutritional Value']}."
    return "I don't have that information."

# Example usage:
user_fruit = 'Apple'
user_question = 'Banana'
response = answer_question(user_fruit, user_question)
print(response)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define the fruits as nodes
fruits = ['Apple', 'Banana', 'Orange']
G.add_nodes_from(fruits)

# Define attributes and values as edges
attributes = {
    'Color': {
        'Apple': ['Red', 'Green'],
        'Banana': 'Yellow',
        'Orange': 'Orange'
    },
    'Taste': {
        'Apple': 'Sweet and slightly tart',
        'Banana': 'Sweet and creamy',
        'Orange': 'Sweet and citrusy'
    },
    'Nutritional Value': {
        'Apple': {
            'Vitamin C': 'Rich',
            'Dietary Fiber': 'High'
        },
        'Banana': {
            'Potassium': 'High',
            'Vitamin B6': 'Rich'
        },
        'Orange': {
            'Vitamin C': 'Excellent'
        }
    }
}

# Add edges for attributes and values
for attribute, values in attributes.items():
    for fruit, value in values.items():
        if isinstance(value, list):
            for v in value:
                G.add_edge(fruit, v, attribute=attribute)
        elif isinstance(value, dict):
            for sub_attribute, sub_value in value.items():
                G.add_edge(fruit, sub_value, attribute=sub_attribute)
        else:
            G.add_edge(fruit, value, attribute=attribute)

# Create a layout for the nodes in the graph
pos = nx.spring_layout(G, seed=42)

# Draw the graph with node labels and edge labels
labels = {node: node for node in G.nodes()}
edge_labels = {(u, v): data['attribute'] for u, v, data in G.edges(data=True)}

plt.figure(figsize=(10, 10))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=5000, node_color='lightblue', font_size=12, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color='red')
plt.title("Semantic Network Representation for Fruits")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define IS-A relation for "Apple" as a fruit
G.add_node("Apple")
G.add_edge("Apple", "Fruit", relation="IS-A")

# Define Kind-of relations for attributes of "Apple"
attributes = {
    'Color': 'Color',
    'Taste': 'Taste',
    'Nutritional Value': 'Nutrient'
}

# Add Kind-of relations for each attribute
for attribute, kind in attributes.items():
    G.add_node(attribute)
    G.add_edge(attribute, kind, relation="Kind-of")

# Create a layout for the nodes in the graph
pos = nx.spring_layout(G, seed=42)

# Draw the graph with node labels and edge labels
labels = {node: node for node in G.nodes()}
edge_labels = {(u, v): data['relation'] for u, v, data in G.edges(data=True)}

plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=5000, node_color='lightblue', font_size=12, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color='red')
plt.title("Semantic Network Representation for 'Apple' with IS-A and Kind-of Relations")
plt.show()
```

```python
# Define the production rules as (condition, action) pairs
production_rules = [
    ({"fruit": "Apple", "question": "Color"}, "Answer: Apple's color is Red and Green."),
    ({"fruit": "Apple", "question": "Taste"}, "Answer: Apple tastes sweet and slightly tart."),
    ({"fruit": "Apple", "question": "Nutritional Value"}, "Answer: Apple is rich in Vitamin C and dietary fiber."),
    ({"fruit": "Banana", "question": "Color"}, "Answer: Banana's color is Yellow."),
    ({"fruit": "Banana", "question": "Taste"}, "Answer: Banana tastes sweet and creamy."),
    ({"fruit": "Banana", "question": "Nutritional Value"}, "Answer: Banana is high in potassium and Vitamin B6."),
    ({"fruit": "Orange", "question": "Color"}, "Answer: Orange's color is Orange."),
    ({"fruit": "Orange", "question": "Taste"}, "Answer: Orange tastes sweet and citrusy."),
    ({"fruit": "Orange", "question": "Nutritional Value"}, "Answer: Orange is an excellent source of Vitamin C."),
    ({"fruit": "Unknown", "question": "Color"}, "Answer: I don't have information about the color of that fruit."),
    ({"fruit": "Unknown", "question": "Taste"}, "Answer: I don't have information about the taste of that fruit."),
    ({"fruit": "Unknown", "question": "Nutritional Value"}, "Answer: I don't have information about the nutritional value of that fruit.")
]

# Function to answer questions based on production rules
def answer_question(fruit, question):
    for condition, action in production_rules:
        if condition["fruit"] == fruit and condition["question"] == question:
            return action
    return "Answer: I don't have that information."

# Example usage:
user_fruit = 'Banana'
user_question = 'Color'
response = answer_question(user_fruit, user_question)
print(response)
```
